```python
import pandas as pd
new = pd.DataFrame()
```

```python
a=[]
for i in range(6,13):
    for j in range(1,32):
        if i==6 and j==1:
            continue
        if i==7 and j<=9:
            continue
        if i==6 and j==31:
            break
        if i==9 and j==31:
            break
        if i==11 and j==31:
            break
        df=pd.read_excel(str(i)+'.'+str(j)+'.xls')
        n,_=df.shape
        #date=df.iloc[n-1].argmax()
        value=df.iloc[n-1].max()
        date=-1
        for k in range(5,29):
            if value==df.iloc[n-1,k]:
                date=k-5
                break
        a.append([date,value])
print(a)
```

```python
df=pd.DataFrame(a)
df
```
