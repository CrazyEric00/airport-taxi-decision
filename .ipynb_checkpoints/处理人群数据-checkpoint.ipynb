```python
import numpy as np
import pandas as pd
import copy
data=pd.read_excel('20190915杭州萧山国际机场.xls')
data=data.sort_values(by='降落时间',axis=0,ascending=True)
val=[[0 for i in range(60)] for j in range(24)]
t=data.iloc[:,3].values
for i in t:
    a,b=i.split(':')
    a=int(a)
    b=int(b)
    na=copy.copy(a)
    nb=copy.copy(b)
    lam = 20
    # 指数分布抽样,size为人数
    add = np.random.exponential(1/lam, size=int(150*0.85))
    ss=0
    count=0
    for i in add:
        ss+=i
        count+=1
        if ss>=1:
            ss=0
            val[na][nb]+=int(count)
            nb+=1
            if nb//60>=1:
                nb%=60
                na+=1
            if na==24:
                break
            count=0
for i in range(24):
    for j in range(60):
        val[i][j]=int(val[i][j]*0.25)
        #print(val[i][j])
      

v = []

f = open('data.txt','w+')
for i in range(24):
    join = val[i]
    v.append(join)
    for j in range(60):
        strNum = str(join[j])
        f.write(strNum)
        if j!=59:
            f.write(' ')
    f.write('\n')
f.close()
```
