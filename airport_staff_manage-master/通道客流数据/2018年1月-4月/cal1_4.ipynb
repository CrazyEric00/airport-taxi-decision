```python
import pandas as pd
new = pd.DataFrame()
```

```python
a=[]
for i in range(1,4):
    for j in range(1,32):
        if i==2 and j==29:
            break
        if i==3 and j==32:
            break
        df=pd.read_excel(str(i)+'.'+str(j)+'.xls')
        n,_=df.shape
        #date=df.iloc[n-1].argmax()
        value=df.iloc[n-1].max()
        date=-1
        for k in range(5,29):
            if value==df.iloc[n-1,k]:
                date=k-5
                break
        a.append([date,value])
print(a)
```

```python
df=pd.DataFrame(a)
df
```

```python
df.to_csv('数据1_4.csv',index=False)
```
